### Eliminamos columnas Unnamed

```python
# Eliminar columnas sin datos
def clear_columns_Unnamed(df):
    # Limpiar espacios en los nombres de las columnas
    df.columns = df.columns.str.strip()

    # Identificar columnas que contienen 'Unnamed'
    columnas_clean = df.filter(like='Unnamed').columns

    # Eliminar esas columnas
    df = df.drop(columns=columnas_clean, errors='ignore')
    
    return df

# Aplicar la función a cada DataFrame en la lista usando un bucle for
for i, df in enumerate(dfs):
    print(f"Procesando DataFrame {i}...")
    dfs[i] = clear_columns_Unnamed(df)
    print(f"Columnas después de limpiar en DataFrame {i}:")
    print(dfs[i].columns)
    print()
```

### Eliminar simbolos de los valores de las columnas de los dfs

```python
def limpiar_simbolos(valor):
    if isinstance(valor, str):
        # Reemplaza los asteriscos por una cadena vacía
        valor = valor.replace('*', '')
        # Elimina espacios adicionales alrededor de las comas
        valor = valor.replace(' ,', ',').replace(', ', ',')
        # Elimina espacios adicionales en general
        valor = valor.strip()
    return valor

def limpiar_df(df):
    # Crea una copia profunda del DataFrame para evitar el SettingWithCopyWarning
    df = df.copy(deep=True)
    # Aplica la función para limpiar asteriscos y ajustar espacios a cada columna
    for col in df.columns:
        df[col] = df[col].apply(lambda x: limpiar_simbolos(x) if pd.notna(x) else x)
    return df
```

### Corregimos la columna de latitud y longitud del dataframe mapa_conectividad

```python
# Eliminar puntos incorrectos y convertir a float
# Esto asume que los puntos están incorrectamente colocados como separadores de miles
mapa_conectividad['latitud'] = mapa_conectividad['latitud'].str.replace('.', '', regex=False).astype(float) / 10**6
mapa_conectividad['longitud'] = mapa_conectividad['longitud'].str.replace('.', '', regex=False).astype(float) / 10**6
```

```python
def corregir_coordenadas(df):
    # Función para corregir la latitud y longitud
    def convertir_coordenada(valor):
        # Eliminar el signo negativo
        valor = abs(valor)
        # Convertir a grados decimales
        return valor / 1e7
    
    # Aplicar la función a las columnas 'latitud' y 'longitud'
    df['latitud'] = df['latitud'].apply(convertir_coordenada)
    df['longitud'] = df['longitud'].apply(convertir_coordenada)
    
    return df
```

```python
# Aplicar la función a tu DataFrame
mapa_conectividad = corregir_coordenadas(mapa_conectividad)
```

### Transformar a formato fecha la columna periodo de los dfs_clean

```python
def transformar_periodo_en_fecha(df):
    # Diccionario para mapear los nombres de los meses a números
    meses = {
        'Ene': 1,
        'Feb': 2,
        'Mar': 3,
        'Abr': 4,
        'May': 5,
        'Jun': 6,
        'Jul': 7,
        'Ago': 8,
        'Sep': 9,
        'Oct': 10,
        'Nov': 11,
        'Dic': 12
    }
    
    # Función auxiliar para extraer fechas de inicio y fin
    def extraer_fechas(periodo):
        try:
            # Extraer los tres primeros caracteres del mes inicial y final
            mes_inicio_abrev, mes_fin_abrev = periodo[:3], periodo[4:7]
            año = periodo[-4:]  # Extraer el año
            
            # Convertir los meses a números
            mes_inicio = meses[mes_inicio_abrev]
            mes_fin = meses[mes_fin_abrev]
            
            # Crear las fechas en formato 'año-mes'
            fecha_inicio = f"{int(año)}-{mes_inicio:02d}"
            fecha_fin = f"{int(año)}-{mes_fin:02d}"
            
            return pd.to_datetime(fecha_inicio), pd.to_datetime(fecha_fin)
        except (ValueError, KeyError, IndexError):
            # Si hay un error en la conversión, devolver NaT
            return pd.NaT, pd.NaT

    # Aplicar la función de extracción a la columna 'periodo'
    df['fecha_inicio'], df['fecha_fin'] = zip(*df['periodo'].apply(extraer_fechas))
    
    return df
```

```python
ingresos = transformar_periodo_en_fecha(ingresos)
penetracion_totales = transformar_periodo_en_fecha(penetracion_totales)
totales_accesos_por_tecnologia = transformar_periodo_en_fecha(totales_accesos_por_tecnologia)
totales_dial_baf = transformar_periodo_en_fecha(totales_dial_baf)
totales_vmd = transformar_periodo_en_fecha(totales_vmd)
```
